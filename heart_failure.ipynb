Step 2 :- Data Understanding

Identify Numericala values & categorical values

```python
num_col=df.select_dtypes(include=['int64','float64']).columns.to_list()
cat_col=df.select_dtypes(include=['object']).columns.to_list()
```

# **Check distribution of numerical features**

```python
df[num_col].hist(bins=20,figsize=(12,8),edgecolor='black',color='teal')
plt.suptitle("Distribtution of numerical features",fontsize=24,fontweight='bold')
plt.show()
```

# **Histograms of Categorical Features**

```python
plt.figure(figsize=(30,20))

for i, feature in enumerate(cat_col):
  plt.subplot(2,3,i+1)
  df[feature].hist()
  plt.title(feature,fontsize=20)
```

Target Distribution

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df,x='HeartDisease')

plt.show()
```

# **Gender vs Heart Disease**

```python
plt.figure(figsize=(6,4))
color=['teal','orange']
sns.countplot(data=df,x='Sex',hue='HeartDisease',palette=color,edgecolor='black')
plt.show()
```

Chest Pain types vs heart Disease

```python
plt.figure(figsize=(6,4))
color=['teal','orange']
sns.countplot(data=df,x='ChestPainType',hue='HeartDisease',palette=color,edgecolor='black')
plt.show()
```
